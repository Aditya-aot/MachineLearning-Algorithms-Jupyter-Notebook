```python
import pandas as pd
df = pd.read_csv("town_area_price.csv")
df
```

```python
dummies=pd.get_dummies(df.torwn)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['torwn','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x= final.drop('price' , axis='columns')
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
# to find accuracy of this nodel
model.score(x,y)
```
