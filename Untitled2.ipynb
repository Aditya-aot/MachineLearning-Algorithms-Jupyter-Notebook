```python
import pandas as pd
import numpy as np

df = pd.read_csv("titanic.csv")
df
```

```python
import math
median_age = math.floor(df.Age.median())
median_age
```

```python
df.Age = df.Age.fillna(median_age)
df
```

```python
x = df.drop(['PassengerId','Name','Embarked','Ticket','Cabin'], axis='columns')
y = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder

le_pclass = LabelEncoder()
le_sex = LabelEncoder()

x['pclass_n']= le_pclass.fit_transform(x['Pclass'])
x['sex_n']= le_sex.fit_transform(x['Sex'])

x.head()
```

```python
x= x.drop(['Pclass', 'Sex'],axis='columns')
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
